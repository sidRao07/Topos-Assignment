## Importing the Packages

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

## The different URL that are being used to scrape the data

```python
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population'
city_url= "https://en.wikipedia.org/wiki/"
crime_url= 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate'
house_url="https://www.kiplinger.com/tool/real-estate/T010-S003-home-prices-in-100-top-u-s-metro-areas/index.php"
income_url = "https://www.forbes.com/sites/jacquelynsmith/2013/11/27/the-u-s-cities-where-people-earn-the-biggest-and-smallest-paychecks/#7a4324283b0d"
cannabis_url = "https://en.wikipedia.org/wiki/Legality_of_cannabis_by_U.S._jurisdiction" 
lifeExpectancy_url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_life_expectancy"
educ_url ="https://en.wikipedia.org/wiki/List_of_U.S._states_by_educational_attainment"
unempRate_url ="https://www.bls.gov/web/metro/laummtrk.htm"
df = pd.DataFrame()
crime_df = pd.DataFrame()
house_df = pd.DataFrame()
income_df = pd.DataFrame()
cannab = pd.DataFrame()
edu_df= pd.DataFrame()
life= pd.DataFrame()
unempRate= pd.DataFrame()

cityUrl= []
```

## Scraper function which will be used to call the request url and return the soup 

```python
def scraperFunc(url):
    response = requests.get(url)       
    soup = BeautifulSoup(response.text,'html.parser')
    return soup
```

## Function to scrape the data from the main wikipedia page United States

Here we use regex to keep only alpha numeric charcters in the dataframe

```python
def mainWikitable(soup):
    wiki_table = soup.find('table',{'class':'wikitable sortable'})
    rank=[]
    city=[]
    state=[]
    estimate_2017=[]
    census_2010=[]
    change =[]
    landAreaMile_2016=[]
    landAreakm2_2016= []
    populationDensityMile_2016=[]
    populationDensityKm2_2016= []
    location=[]
    latitude=[]
    longitude=[]    
    x=0
    for row in wiki_table.findAll('tr'):
        if x<11 :
            x+=1
            cells=row.findAll('td')
            #print(len(cells))
            if len(cells)==11:
                rank.append(cells[0].find(text=True))
                city.append(cells[1].find(text=True))
                state.append(cells[2].findAll(text=True)[1])
                estimate_2017.append(cells[3].find(text=True))  
                census_2010.append(cells[4].find(text=True))
                change.append(cells[5].find(text=True))
                landAreaMile_2016.append(cells[6].find(text=True))
                landAreakm2_2016.append(cells[7].find(text=True))
                populationDensityMile_2016.append(cells[8].find(text=True))
                populationDensityKm2_2016.append(cells[9].find(text=True))
                location.append(cells[10].findAll(text=True)[4])
                latitude.append(cells[10].findAll(text=True)[4].split(" ")[0])
                longitude.append(cells[10].findAll(text=True)[4].split(" ")[1])
                
    rank = [re.sub(r"[^a-zA-Z0-9]", "", file) for file in rank]
    state = [re.sub(r"[^a-zA-Z0-9]", "", file) for file in state]
    estimate_2017 = [re.sub(r"[^a-zA-Z0-9]", "", file) for file in estimate_2017]
    census_2010 = [re.sub(r"[^a-zA-Z0-9]", "", file) for file in census_2010]
    landAreaMile_2016 = [re.sub(r"[^0-9]", "", file) for file in landAreaMile_2016]
    landAreakm2_2016 = [re.sub(r"[^0-9]", "", file) for file in landAreakm2_2016]
    populationDensityMile_2016 = [re.sub(r"[^0-9]", "", file) for file in populationDensityMile_2016]
    populationDensityKm2_2016 = [re.sub(r"[^0-9]", "", file) for file in populationDensityKm2_2016] 
    latitude = [re.sub(r"[^a-zA-Z0-9.]", "", file) for file in latitude] 
    longitude = [re.sub(r"[^a-zA-Z0-9.]", "", file) for file in longitude] 
    change = [i.replace(u'\u2212', '-') for i in change]
    change = [re.sub(r"[^0-9]", "", file) for file in change]
    
    df['population rank 2018'] = rank
    df['city'] = city
    df['state'] = state
    df['estimate_2017'] = estimate_2017
    df['census_2010'] = census_2010
    df['change'] = change
    df['landAreaMile_2016'] = landAreaMile_2016
    df['landAreakm2_2016'] = landAreakm2_2016
    df['populationDensityMile_2016'] = populationDensityMile_2016
    df['populationDensityKm2_2016'] = populationDensityKm2_2016
    #df['location'] = location
    df['latitude'] = latitude
    df['longitude'] = longitude
    return city
```

```python
soup= scraperFunc(url)
city= mainWikitable(soup)
```

## Forming the different cities url(wikipedia page for each city)
For the cities that have spacein the, we replace the space with underscore(_) and append them to the city url

```python
def formUrl():
    for i in city:
        i=i.replace(" ", "_")
        if i == "Phoenix":
            i = "Phoenix_Arizona"
        elif i== "San_Jose" :
            i = "San_Jose,_California"
        elif i== "Columbus" :
            i = "Columbus,_Ohio"    
        cityUrl.append(city_url+i)
        
```

## Scraping data from the individual city pages
table1 contains the demographic data, while table2 gets data from below the city picture on the right side of the page

```python
white=[]
nonHisp=[]
africanAmerican=[]
hispanic=[]
asian=[]
mayor_name=[]
time_zone= []
def cityWikipages():
    x=0
    for i in cityUrl:
        soup = scraperFunc(i)
        table1= soup.find('table',{'class':'wikitable sortable collapsible'})
        
        x+=1
        for row in table1.findAll('tr'):
            cells=row.findAll('td')
            
            if len(cells)==5:
                
                
                if cells[0].find(text=True) == "White" or cells[0].find(text=True)== "White or European American" or cells[0].find(text=True) =="Non-Hispanic White":
                    white.append(cells[1].find(text=True))
                elif cells[0].find(text=True) == "Black or African American":
                    africanAmerican.append(cells[1].find(text=True))
                elif cells[0].find(text=True) =="Hispanic or Latino":
                    if(cells[1].find(text=True)!= "n/a"):
                        hispanic.append(cells[1].find(text=True))
                elif cells[0].find(text=True) == "Asian" :
                    asian.append(cells[1].find(text=True))
              
        if x!= len(white):
                white.append('0')
        elif x!= len(africanAmerican):
                africanAmerican.append('0')
        elif x!= len(hispanic):
                hispanic.append('0')
        elif x!= len(asian):
                asian.append('0')
        
        table2 =soup.find('table',{'class':'infobox geography vcard'})
        for row in table2.findAll('tr'):
            if row.find('th') is not None and "Mayor" in row.find('th').text.strip():
                mayor_name.append(row.find('td').text.strip().split("[")[0])
                
            if row.find('th') is not None and "Time zone" in row.find('th').text.strip():
                time_zone.append(row.find('td').text.strip())    

```

Regex is used to make sure only alphanumeric characters are present in the dataframe

```python
formUrl()
cityWikipages()

time_zone = [i.replace(u'\u2212', '-') for i in time_zone]
white = [re.sub(r"[^a-zA-Z0-9.]", "", file) for file in white]
africanAmerican = [re.sub(r"[^a-zA-Z0-9.]", "", file) for file in africanAmerican]
hispanic = [re.sub(r"[^a-zA-Z0-9.]", "", file) for file in hispanic]
asian = [re.sub(r"[^a-zA-Z0-9.]", "", file) for file in asian]

df['white'] = white
df['africanAmerican'] = africanAmerican
df['hispanic'] = hispanic
df['asian'] = asian
df['Mayor Name'] = mayor_name
df['Time Zone'] = time_zone
```

## Scraping Data for crime rates in different cities

```python
city1 = []
state1 = []
violentCrime = []
propertyCrime = []

def crimeWikiTable(soup):
    crime_table = soup.find('table',{'class':'wikitable sortable'})

    for row in crime_table.findAll('tr'):
       
        cells=row.findAll('td')
      #  print(len(cells))
        if len(cells) ==13 :
            if cells[1].find(text=True)=="New York" :
                city1.append("New York City")
            else :
                city1.append(cells[1].find(text=True))
            if cells[0].find(text=True)=="New York":
                state1.append("NewYork")
            else:   
                state1.append(cells[0].find(text=True))
            violentCrime.append(cells[3].find(text=True))
            propertyCrime.append(cells[8].find(text=True))      
```

```python

soup= scraperFunc(crime_url)
crimeWikiTable(soup)
```

```python
crime_df['city'] = city1
crime_df['state'] = state1
crime_df['Violent Crime'] = violentCrime
crime_df['propertyCrime'] = propertyCrime
#crime_df   
```

we merge the 2 dataframes on city and state

```python
df = pd.merge(df, crime_df, how='inner', on=['city', 'state'])
```

## Scraping Data to get the houses prices in different cities

```python
city1=[]
price =[]
def houseRate():
    
    house_table = soup.find('table',{'class':'kip-table sortable'})

    for row in house_table.findAll('tr'):

            cells=row.findAll('td')
            if len(cells) ==6 :
                x= cells[0].find(text=True).split(", ")
                if x[0]=="New York" :
                    city1.append("New York City")
                else :
                    city1.append(x[0])
               
                price.append(cells[1].find(text=True))
```

```python
soup= scraperFunc(house_url)
houseRate()
price = [re.sub(r"[^0-9]", "", file) for file in price]
house_df['city'] = city1
house_df['house price'] = price
```

```python
df = pd.merge(df, house_df, how='inner', on=['city'])
```

## Scraping Data to get the per capita income of the people in different cities

```python
city1=[]
income =[]
def perCapitaIncome():
    income_table = soup.find('table',)

    for row in income_table.findAll('tr'):

        cells=row.findAll('td')
#        print(len(cells))
        if len(cells) == 4:
            if cells[0].find(text=True).split("-")[0]=="New York" :
                city1.append("New York City")
            elif cells[0].find(text=True).split("-")[0]=="San Antonio, TX" :
                city1.append("San Antonio")
            else :
                city1.append(cells[0].find(text=True).split("-")[0])
            income.append(cells[1].find(text=True)) 
```

```python
soup= scraperFunc(income_url)
perCapitaIncome()
income = [re.sub(r"[^0-9]", "", file) for file in income]
income_df['city'] = city1
income_df['per capita income'] = income
```

```python
df = pd.merge(df, income_df, how='inner', on=['city'])
```

## Scraping Data to check if Marijuana is legal or illegal in the city

```python
state1=[]
recr_canab=[]

def cannabisLegality():
    
    can_table = soup.find('table',{'class':'wikitable sortable'})
    for row in can_table.findAll('tr'):

        cells=row.findAll('td')

        if len(cells) ==7 :
            if cells[0].find(title=True).text.strip()=="New York" :
                state1.append("NewYork")
            else:    
                state1.append(cells[0].find(title=True).text.strip())
            recr_canab.append(cells[2].find(text=True).split(" ")[0])
```

```python
soup= scraperFunc(cannabis_url)
cannabisLegality()

cannab['state'] = state1
recr_canab = [re.sub(r"[^a-zA-Z0-9 ]", "", file) for file in recr_canab] 
cannab['recreational Cannabis'] = recr_canab

```

```python
df = pd.merge(df, cannab, how='left', on=['state'])
```

## Scraping Data to find the unemployement rate in different cities

```python
city1=[]
unemp=[]

def unemeploymentRate():
    unemp_table = soup.find('table',{'class':'regular'})


    for row in unemp_table.findAll('tr'):

        cell1=row.findAll('th')
        cells=row.findAll('td')
       # print(len(cells))
        if len(cell1)==1 :
            if cell1[0].find(text=True).split("-")[0]=="New York" :
                city1.append("New York City")
            else :
                city1.append(cell1[0].find(text=True).split("-")[0])

        if len(cells) == 2:
            unemp.append(cells[1].find(text=True))


    
```

```python
soup= scraperFunc(unempRate_url)
unemeploymentRate()

unempRate['city'] = city1
unempRate['Unemployment  Rank']= unemp
unempRate
df = pd.merge(df, unempRate, how='inner', on=['city'])
```
